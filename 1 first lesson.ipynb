```python
import numpy as np
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(2,2)
```

```python
# INsert an Hadamart gate
qc.h(0)
```

```python
#Insert a control not (0 is the control and 1 is the target)
qc.cx(0,1)
```

```python
#Measure both of them
qc.measure([0,1],[0,1])
```

```python
qc.draw('mpl')
```

```python
from qiskit.providers.aer import QasmSimulator
```

```python
from qiskit import transpile
```

```python
simulator = QasmSimulator()
compiled_qc = transpile(qc,)
```

```python
#Job is the result of the simulation of your circuit
job = simulator.run(compiled_qc, shots=10000)

# The ral computer will make the measure only at the end. To see the histograms which are done statistically (not analitycally)
```

```python
result = job.result()
counts = result.get_counts(compiled_qc)
```

```python
#dictionary of the result
counts
```

```python
from qiskit.visualization import plot_histogram
```

```python
#Take dictionary as a parameter
plot_histogram(counts)
```
